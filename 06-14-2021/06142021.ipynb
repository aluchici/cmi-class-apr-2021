```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import graphviz 
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# train DT classifier
X = [[0, 0], [1, 1], [2, 2], [1.5, 2.5], [1.7, 2.3]]
Y = ['class a', 'class b', 'class c', 'class b', 'class c']
clf = DecisionTreeClassifier(criterion='entropy')
# split data into train & test
X_train = X[:2]
Y_train = Y[:2]
X_test = X[3:]
Y_test = Y[3:]
# cross validation
clf = clf.fit(X_train, Y_train)
```

```python
# test model
Y_predict = clf.predict(X_test)
acc = accuracy_score(Y_test, Y_predict)
print('Accuracy: %s' % acc)
```

```python
iris = load_iris()
X, Y = iris.data, iris.target
X_train = X[:int(len(X)*0.75)]
X_test = X[int(len(X)*0.75)+1:]
Y_train = Y[:int(len(X)*0.75)]
Y_test = Y[int(len(X)*0.75)+1:]
clf_iris = DecisionTreeClassifier(criterion='entropy', min_samples_leaf=4)
clf_iris = clf_iris.fit(X_train, Y_train)
```

```python
fig = plt.figure(figsize=(20, 10))
_ = tree.plot_tree(clf_iris)
plt.show()
```

```python
Y_predict = clf_iris.predict(X_test)
acc = accuracy_score(Y_test, Y_predict)
print("Accuracy: %s" % acc)
```
