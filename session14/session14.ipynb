```python
from scipy.stats import binom, geom
import matplotlib.pyplot as plt
import numpy as np 
n, p = 10, 0.5
mean, var = binom.stats(n, p, moments='mv')
print('Mean: %s, Variance: %s' % (mean, var))
```

```python
p = 0.3
mean, var = geom.stats(p, moments='mv')
print('Mean: %s, Var: %s' % (mean, var))
```

```python
import pandas as pd 
from scipy.stats import expon, rv_continuous

data = pd.read_csv('../session12/vgsales.csv')
data = data[data['Platform'] == 'X360']['Global_Sales']

# fit dist to data
params = expon.fit(data)

# Separate parts of parameters
arg = params[:-2]
loc = params[-2]
scale = params[-1]

# Calculate fitted PDF and error with fit in distribution
pdf = expon.pdf(data, loc=loc, scale=scale, *arg)
```

```python
plt.figure(figsize=(20, 10))
plt.hist(data, bins=25)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data, pdf)
plt.show()
```
